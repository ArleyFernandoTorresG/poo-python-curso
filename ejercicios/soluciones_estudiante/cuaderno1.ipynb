```python
import pandas as pd
```

```python
def primos(n):
    """Genera una lista de números primos hasta n."""
    primos = []
    for num in range(2, n + 1):
        es_primo = True
        for divisor in range(2, int(num**0.5) + 1):
            if num % divisor == 0:
                es_primo = False
                break
        if es_primo:
            primos.append(num)
    return primos
```

```python
for i in [2,3,5,8,23]:
    print(primos(i))
```
